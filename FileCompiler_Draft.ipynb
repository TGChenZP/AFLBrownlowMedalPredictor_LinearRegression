```python
import pandas as pd
import os
```

```python
choice = 'OriginalData'
# choice = 'ManipulatedData'
# choice = 'StandardisedData'
```

```python
filelist = os.listdir(f'./Data (second scrape) - fixed/{choice}')[1:]
# Remove the first file (an ipynb checkpoint file)
```

Open all the files and concat them to the main file

```python
out = pd.read_csv(f'./Data (second scrape) - fixed/{choice}/{filelist[0]}')
for file in filelist[1:]:
    df = pd.read_csv(f'./Data (second scrape) - fixed/{choice}/{file}')
    out = pd.concat([out, df], axis=0)
```

```python
out.index = range(0,len(out))
```
