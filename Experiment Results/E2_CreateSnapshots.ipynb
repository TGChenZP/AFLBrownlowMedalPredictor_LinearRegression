**0. Import Libraries**

```python
import pandas as pd
```

```python
file = pd.read_csv('./Groupby Output/All.csv')
```

**1. Create Snapshots**

```python
# Snapshot based on average correlation coefficient
a = file.sort_values(['coef_avg'], ascending = [False])
a.to_csv('all_coef.csv', index = False)
```

```python
# Snapshot based on tp3 (and then tp2, tp1)
b = file.sort_values(['tp3', 'tp2', 'tp1'], ascending = [False, False, False])
b.to_csv('all_tp.csv', index = False)
```

```python
# Snapshot based on top m1_Top3 score (and secondary least m1_avgvotediff(3))
c = file.sort_values(['m1_Top3score', 'm1_avgvotediff(3)'], ascending = [False, True])
c.to_csv('all_m1_3.csv', index = False)
```

```python
# Snapshot based on top m1_Top4 score (and secondary least m1_avgvotediff(4))
d = file.sort_values(['m1_Top4score', 'm1_avgvotediff(4)'], ascending = [False, True])
d.to_csv('all_m1_4.csv', index = False)
```

```python
# Snapshot based on top m2_Winner score (and secondary least m2_avgvotediff)
e = file.sort_values(['m2_Winner', 'm2_avgvotediff'], ascending = [False, True])
e.to_csv('all_m2.csv', index = False)
```
